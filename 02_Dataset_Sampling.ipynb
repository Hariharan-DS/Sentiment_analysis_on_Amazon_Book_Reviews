# Preprocessed Dataset size before sampling 

```python
book_df.shape
```

```python
# Step 1: Exclude 'Fiction' category
filtered_data_no_fiction = book_df[book_df['categories'] != "Fiction"]

# Step 2: Identify top 10 categories excluding 'Fiction'
top_categories_no_fiction = (
    filtered_data_no_fiction.groupby('categories')
    .size()
    .sort_values(ascending=False)
    .head(10)
    .index
)

# Step 3: Filter data for the new top 10 categories
filtered_data_final = filtered_data_no_fiction[filtered_data_no_fiction['categories'].isin(top_categories_no_fiction)]

# Step 4: Bin the continuous 'review/score' into intervals
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']
filtered_data_final['score_bins'] = pd.cut(filtered_data_final['review/score'], bins=bins, labels=labels, right=False)

# Step 5: Grouping data by categories and score bins
score_distribution = filtered_data_final.groupby(['categories', 'score_bins']).size().reset_index(name='count')

# Normalize the distribution within each category (optional)
score_distribution['normalized_count'] = score_distribution.groupby('categories')['count'].transform(lambda x: x)

# Step 6: Plotting the distribution
plt.figure(figsize=(14, 8))
sns.barplot(
    x='categories',
    y='normalized_count',
    hue='score_bins',
    data=score_distribution,
    palette='viridis'
)

# Customize the plot
plt.title('Distribution of Review Scores by Top 10 Categories (Excluding Fiction)', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Count of Review Scores', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Score Bins', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Define the desired sample size
sample_size = 40000

# Calculate the sample size per group based on the overall distribution of review score and category in the data
# This applies a two-level stratification
sampled_data = book_df.groupby(['categories', 'review/score'], group_keys=False).apply(
    lambda x: x.sample(int(sample_size * len(x) / len(book_df)), random_state=42)
)

# Save the sample to a new CSV or use it directly for training
sampled_data.to_csv(r"C:\Users\HAN4COB\.conda\envs\test\Data_Science_in_practice/sampled_book_df.csv", index=False)
```
